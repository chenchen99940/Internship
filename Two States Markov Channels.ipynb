```python
import numpy as np
import matplotlib.pyplot as plt
p=0.8;
r=0.2;
def a(l):
    return (r-r*(p-r)**l)/(1+r-p);
def b(l):
    return (r+(1-p)*(p-r)**l)/(1+r-p);

def whittle(l):
    return ((a(l)-a(l+1))*(l+1)+a(l+1))/(1-p+(a(l)-a(l+1))*l+a(l+1));

x = np.arange(0, 25, 1);
y = a(x);
z = b(x);
w=whittle(x);

plt.subplot(1,2,1);
plt.plot(x, y);

plt.subplot(1,2,1);
plt.plot(x,z);

plt.subplot(1,2,2);
plt.plot(x,w);

plt.subplot(1,2,2);
plt.axhline(y= r/((1-p)*(1+r-p)+r), color='r', linestyle='-')
```
